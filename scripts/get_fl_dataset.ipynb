```python
import os
import re
import random
```

```python
# settings
client_num = 10
train_pro = 0.2
split = "28"
tag_prefix = "bal_iid"
```

```python
# get iid dataset
dst_pth = "../Dataset/nwpu_resisc45/"
all_img_pth = os.path.join(dst_pth, "all_img")
img_names = os.listdir(all_img_pth)
ids = list(range(len(img_names)))
random.shuffle(ids)
per_client_ids = []
per_client_num = int(len(img_names) / client_num)
for i in range(client_num):
    per_client_ids.append(ids[i * per_client_num: (i + 1) * per_client_num])

# map class name to number
cls_names = ['airplane', 'airport', 'baseball_diamond', 'basketball_court', 'beach', 'bridge', 'chaparral', 'church', 'circular_farmland', 'cloud', 'commercial_area', 'dense_residential', 'desert', 'forest', 'freeway', 'golf_course', 'ground_track_field', 'harbor', 'industrial_area', 'intersection', 'island', 'lake', 'meadow', 'medium_residential', 'mobile_home_park', 'mountain', 'overpass', 'palace', 'parking_lot', 'railway', 'railway_station', 'rectangular_farmland', 'river', 'roundabout', 'runway', 'sea_ice', 'ship', 'snowberg', 'sparse_residential', 'stadium', 'storage_tank', 'tennis_court', 'terrace', 'thermal_power_station', 'wetland']
cls2num = {}
for i in range(len(cls_names)):
    cls2num[cls_names[i]] = i
print("cls2num: ", cls2num)

for i in range(client_num):
    # split every client datasets into two parts, including training dataset and valid dataset
    cur_client_ids = per_client_ids[i]
    random.shuffle(cur_client_ids)
    cur_client_train_num = int(len(cur_client_ids) * train_pro)
    cur_client_train_ids = cur_client_ids[:cur_client_train_num]
    cur_client_valid_ids = cur_client_ids[cur_client_train_num:]
    # write, tag is `fl + client id`
    with open(os.path.join(dst_pth, 'train_labels_{}_{}{}.txt'.format(split, tag_prefix, i)), "w") as f:        
        for id in cur_client_train_ids:
            img_name = img_names[id]
            cls_name = re.search("\D*\d", img_name).span()
            cls_name = img_name[cls_name[0]: cls_name[1]][:-2]
            cls_name = str(cls_name)
            f.writelines([f"{img_name} {cls_name} {cls2num[cls_name]}\n", ])
    with open(os.path.join(dst_pth, 'valid_labels_{}_{}{}.txt'.format(split, tag_prefix, i)), "w") as f:        
        for id in cur_client_valid_ids:
            img_name = img_names[id]
            cls_name = re.search("\D*\d", img_name).span()
            cls_name = img_name[cls_name[0]: cls_name[1]][:-2]
            cls_name = str(cls_name)
            f.writelines([f"{img_name} {cls_name} {cls2num[cls_name]}\n", ])
```
